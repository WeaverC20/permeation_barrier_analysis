```python
def load_TC_data(filepath):
    """
    Loads CSV with a datetime RealTimestamp column and returns:
    - time_s : seconds since start (numpy array)
    - V_Baratron : downstream baratron voltage (numpy array)
    - V_Wasp : wasp voltage (numpy array)
    - df : cleaned DataFrame with Python-safe column names
    """
    # Read CSV
    df = pd.read_csv(filepath)

    # Parse datetimes
    df["RealTimestamp"] = pd.to_datetime(df["RealTimestamp"], errors="coerce")

    # Time since start (0 = first row)
    t0 = df["RealTimestamp"].iloc[0]
    time_s = (df["RealTimestamp"] - t0).dt.total_seconds().to_numpy()

    # Rename columns to be Python-friendly
    df = df.rename(columns=lambda x: x.strip().replace(" ", "_").replace("(", "").replace(")", "").replace("/", "_per_"))

    return time_s, df

T_t, T_df = load_TC_data(data_path + "/thermocouple_data.csv")

print(T_df.columns)

try:
    T_data = T_df["type_K_thermocouple_Voltage_mV"].to_numpy(dtype=float)
except KeyError:
    T_data = T_df["type K thermocouple_Voltage (mV)"].to_numpy(dtype=float)

print(T_data)

T_data = voltage_to_temp_typeK(T_data)

temperature_C = np.mean(T_data[int(len(T_data)/4):])

temperature_K = temperature_C + 273.15

# for old runs
# temperature_K = temperature_C_set + 273.15 - 18.11
# temperature_K = temperature_C_set + 273.15 - 19.62 # for 500C runs
```

```python

try:
    from scipy.signal import savgol_filter
except ImportError:
    savgol_filter = None

def smooth_derivative_savgol(y, dx, window=11, poly=3):
    """
    Smoothed first derivative using Savitzky–Golay.
    window: odd int (# of samples), increase to smooth more
    poly: polynomial order (2–4 usually)
    """
    if savgol_filter is None:
        raise ImportError("scipy is required for Savitzky–Golay (pip install scipy).")
    if window % 2 == 0:
        window += 1
    return savgol_filter(y, window_length=window, polyorder=poly, deriv=1, delta=dx, mode="interp")

dPdt = np.gradient(P_Baratron_downstream, t)

dPdt_smooth = smooth_derivative_savgol(P_Baratron_downstream, dx=t[1]-t[0], window=155, poly=3)

plt.figure(figsize=(7,5))

# Raw (noisy)
plt.plot(t, dPdt, label="Raw dP/dt", linewidth=1, alpha=0.6)

# Smoothed (on top)
plt.plot(t, dPdt_smooth, label="Smoothed dP/dt", linewidth=2)

plt.xlabel("Time (s)")
plt.ylabel("dP/dt (Torr/s)")
plt.title("Rate of Pressure Change Downstream")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Using timelag to calculate D_eff
tau_l = x_cross - t_0
D_eff = e**2 / (6 * tau_l)
phi_timelag = slope*133.3 / (R*temperature_K/V*(P_up*133.3)**0.5*e*Area) * 6.022*10**23

print(f"Effective diffusion coefficient: D_eff = {D_eff:.2e} m^2/s")
print(f"Permeability from timelag: phi = {phi_timelag:.2e} mol/(m·s·Pa^0.5)")

print(slope *3600)
```

```python
# Using Flux to calculate permeability

dPdt = dPdt_smooth * 133.3 # convert Torr/s to Pa/s
J = dPdt*V / (R*temperature_K*Area) * 6.022*10**23 # H/(m^2*s)

Perm = J*e / (P_up*133.3)**0.5

# print(f"Permeability (continuum calculation): Phi = {Perm:.2e}")
```

```python
# Takaishi-Sensui approximation for the Thermal Transpiration Effect correction

V1_ratio = ufloat(0.35, 0.1) # ratio of V1 to total volume
# V1_ratio = 0.35

V1 = V*V1_ratio
V2 = V*(1-V1_ratio)
T1 = temperature_K
T2 = 300 # ambient temperature in Kelvin

A  = 1.24 * 56.3 / 10e-5
B  = 8 * 7.7 / 10e-2
C  = 10.6 * 2.73
d = 0.0155 # diameter of pipe

P2dot = dPdt
P2 = P_Wasp_downstream[-1] * 133.3 # convert Torr to Pa


# --- helper quantities ---
num2 = C*(d*P2)**0.5 + (T2/T1)**0.5 + A*d**2*P2**2 + B*d*P2     # #2
den3 = C*(d*P2)**0.5 + A*d**2*P2**2 + B*d*P2 + 1                # #3
num1 = B*d*P2dot + (C*d*P2dot)/(2*(d*P2)**0.5) + 2*A*d**2*P2*P2dot  # #1

# --- assemble dn/dt ---
n_dot = (V2*P2dot)/(R*T2) \
      + (V1*P2dot)/(R*T1*num2) \
      + (V1*P2*num1)/(R*T1*num2*den3) \
      - (V1*P2*num1)/(R*T1*num2**2)

J_TS = n_dot / Area * 6.022*10**23 # H/(m^2*s)

Perm_TS = J_TS*e / (P_up*133.3)**0.5 # Permeability from Takaishi-Sensui
```

```python
# free molecular flow approximation for the Thermal Transpiration Effect correction

dndt = dPdt * (V1/(R*(T2*T1)**0.5) + V2/(R*T2)) # dn/dt from free molecular flow approx

J_FMF = dndt / Area * 6.022*10**23 # H/(m^2*s)

Perm_FMF = J_FMF*e / (P_up*133.3)**0.5 # Permeability from FMF
```

```python
Perm_nom = unp.nominal_values(Perm)
Perm_TS_nom = unp.nominal_values(Perm_TS)
Perm_FMF_nom = unp.nominal_values(Perm_FMF)

plt.figure()
plt.plot(t, Perm_nom, label="Permeability Continuum Flow")
plt.plot(t, Perm_TS_nom, label="Permeability Takaishi-Sensui")
plt.plot(t, Perm_FMF_nom, label="Permeability Free Molecular Flow")
plt.xlabel("Permeability (mol/(m·s·Pa^0.5))")
plt.ylabel("Time (s)")
plt.title("Permeability vs Time")
plt.grid(True)
plt.legend()
plt.show()

print("Max TS Perm:", np.max(Perm_TS))
print("Min TS Perm:", Perm_TS[-1])
```

```python
def find_start_after_spike(x, y, num_points=10, step=20, threshold_factor=5):
    """
    Linearly fits number of points to fidn location where slope of spike is
    below the threshod factor times the average slope after the spike.

    Returns: i, the index in x,y after the spike which is in the middle of the
    num_points threshold points used for fitting.
    """
    m_rest, b_rest = np.polyfit(x[round(len(x)*3/4):], y[round(len(x)*3/4):], 1)

    found_spike = False
    i = 0
    while not found_spike:
        try:
            m, b = np.polyfit(x[i*step:((i+1)*step+num_points)], y[i*step:((i+1)*step+num_points)], 1)
            if abs(m) < threshold_factor * abs(m_rest):
                found_spike = True
                return(i*step + round(num_points/2))
            else:
                i += 1
        except KeyError:
            return(round(num_points/2))
```

```python
perm_TS_std = unp.std_devs(Perm_TS)

# Nominal permeability values and their uncertainties
idx = np.argmax(Perm_TS_nom)

# To offset a little from the spike
idx2 = find_start_after_spike(t[idx:], Perm_TS_nom[idx:])

P = np.array([Perm_TS_nom[idx+idx2], Perm_TS_nom[-1]])
sigma = np.array([perm_TS_std[idx+idx2], perm_TS_std[-1]])

# ---- STEP 1: Weighted Mean Plateau Value ----
weights = 1 / sigma**2
P_avg = np.sum(P * weights) / np.sum(weights)

# ---- STEP 2: Combine measurement + data scatter ----
sigma_meas = np.mean(sigma)              # average measurement uncertainty
sigma_scatter = np.std(P - P_avg)        # scatter around mean
sigma_total = np.sqrt(sigma_meas**2 + sigma_scatter**2)

# ---- STEP 3: Uncertainty of the mean ----
sigma_avg = sigma_total / np.sqrt(len(P))

# ---- STEP 4: Upper / Lower bounds ----
P_high = P_avg + sigma_avg
P_low  = P_avg - sigma_avg

print(f"Plateau permeability: {P_avg:.3e} ± {sigma_avg:.3e}")
print(f"Upper bound: {P_high:.3e}")
print(f"Lower bound: {P_low:.3e}")
```

```python
plt.figure(figsize=(7,5))

# Original permeability curves
plt.plot(t, Perm_nom, label="Permeability Continuum Flow")
plt.plot(t, Perm_TS_nom, label="Permeability Takaishi-Sensui")
plt.plot(t, Perm_FMF_nom, label="Permeability Free Molecular Flow")

# --- Horizontal plateau line ---
plt.hlines(P_avg, t[0], t[-1], color='black', linestyle='--', linewidth=2,
           label=f"Plateau Mean = {P_avg:.2e}")

# --- Shaded ± uncertainty band ---
plt.fill_between(t, P_low, P_high, color='gray', alpha=0.25,
                 label=f"±1σ = {sigma_avg:.2e}")

plt.xlabel("Time (s)")
plt.ylabel("Permeability (mol/(m·s·Pa$^{0.5}$))")
plt.title("Permeability vs Time with Plateau Fit")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
